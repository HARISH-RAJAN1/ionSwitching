```python
import numpy as np
import pandas as pd
import time 
import os
```

```python
os.getcwd()
```

```python
dataset=pd.read_csv("train.csv")
```

```python
def batchnum(ts,div):
  qt=(float(ts)*10000)//div
  if ((float(ts)*10000)%div)==0:
    return 'b'+str(int(qt))
  else:
    return 'b'+str(int(qt+1))
```

```python
dataset['batch_discrete']=dataset.time.apply(lambda x : batchnum(x,500000))
dataset['batch_sec']=dataset.time.apply(lambda x : batchnum(x,10000))
dataset['batch_10khz']=dataset.time.apply(lambda x : batchnum(x,500))
```

```python
for param in ['batch_discrete','batch_sec','batch_10khz']:
    dct = {}
    dct['mean'+param] = dataset.groupby([param])['signal'].mean()
    dct['median'+param] = dataset.groupby([param])['signal'].median()
    dct['max'+param] = dataset.groupby([param])['signal'].max()
    dct['min'+param] = dataset.groupby([param]) ['signal'].min()
    dct['std'+param] = dataset.groupby([param]) ['signal'].std()
    dct['25thpercentile'+param] = dataset.groupby([param]) ['signal'].quantile(.25)
    dct['75thpercentile'+param] = dataset.groupby([param]) ['signal'].quantile(.75)
    for val in dct:
        dataset[val] = dataset[param].map(dct[val].to_dict())
    dataset['range'+param]=dataset['max'+param]-dataset['min'+param]
    dataset['minmax'+param]=dataset['min'+param]/dataset['max'+param]
    dataset['prev'+param]=dataset['signal'].shift(-1).fillna(dataset['signal'].values[-2])
    dataset['next'+param]=dataset['signal'].shift(+1).fillna(dataset['signal'].values[1])
```

```python
dataset
```

```python
X=dataset.drop(columns=['time','open_channels','batch_discrete',  'batch_sec','batch_10khz','signal']).values
y=dataset["open_channels"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```
